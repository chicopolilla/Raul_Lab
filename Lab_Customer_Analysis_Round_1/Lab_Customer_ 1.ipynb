```python
#Delete the column education and the number of open complaints from the dataframe.
file_todos =file_todos.drop(columns=['Education','Number of Open Complaints'],axis=1)
file_todos
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
file_todos['Customer Lifetime Value']= file_todos['Customer Lifetime Value'].replace({'%':''}, regex=True)
file_todos
```

```python
file_todos['Customer Lifetime Value'] = file_todos['Customer Lifetime Value']*100

```

```python
file_todos
```

```python
file_todos['Customer Lifetime Value'] = pd.to_numeric(file_todos['Customer Lifetime Value'],
                                                      errors= 'coerce')
file_todos
```
